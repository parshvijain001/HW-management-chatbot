```python
import json
import os
from textblob import TextBlob

def load_tasks():
    if os.path.exists("homework_data.json"):
        with open("homework_data.json", "r") as file:
            try:
                return json.load(file)
            except:
                return []
    else:
        return []

def save_tasks(tasks):
    with open("homework_data.json", "w") as file:
        json.dump(tasks, file)

# --- Bot Features ---

def add_task(sentence):
    tasks = load_tasks()
    tasks.append({"task": sentence, "done": False})
    save_tasks(tasks)
    print("‚úÖ Task added!")

def view_tasks():
    tasks = load_tasks()
    if len(tasks) == 0:
        print("üì≠ No homework tasks yet.")
    else:
        i = 1
        for task in tasks:
            status = "‚úì Done" if task["done"] else "‚ùå Not done"
            print(f"{i}) {task['task']} - {status}")
            i += 1

def mark_done(number):
    tasks = load_tasks()
    index = number - 1
    if 0 <= index < len(tasks):
        tasks[index]["done"] = True
        save_tasks(tasks)
        print("‚úÖ Task marked as done!")
    else:
        print("‚ùå Invalid task number.")

def get_intent(sentence):
    sentence = sentence.lower()
    if "add" in sentence or "need to do" in sentence or "homework" in sentence:
        return "add"
    elif "show" in sentence or "view" in sentence or "what is" in sentence:
        return "view"
    elif "done" in sentence or "finished" in sentence or "complete" in sentence:
        return "mark"
    elif "exit" in sentence or "quit" in sentence:
        return "exit"
    else:
        return "unknown"
# --- Main Chatbot Loop ---

def chatbot():
    print("üìö Welcome to the NLP Homework Bot!")
    print("You can type things like:")
    print(" - I need to do my science homework")
    print(" - Show my homework")
    print(" - I finished task 1")
    print(" - Exit\n")

    while True:
        user_input = input("You: ")
        intent = get_intent(user_input)

        if intent == "add":
            add_task(user_input)
        elif intent == "view":
            view_tasks()
        elif intent == "mark":
            words = user_input.split()
            for word in words:
                if word.isdigit():
                    mark_done(int(word))
                    break
            else:
                print("‚ùì Please mention which task number you finished.")
        elif intent == "exit":
            print("üëã Bye! Good luck with your homework.")
            break
        else:
            print("‚ùì I didn‚Äôt understand that. Try again.")


chatbot()
```
